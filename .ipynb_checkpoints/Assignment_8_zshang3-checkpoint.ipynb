````python
# Data load
df = pd.read_csv('~/Desktop/CS4414/energy_appliances_standard.csv')
print(df.head())
````

````python
# Create train / test split
train_data,test_data = train_test_split(df,test_size=0.3,random_state = 20201107)
````

````python
# Define random forest and iterate over the number of trees
X = train_data.drop(['Appliances'], axis='columns').values
y = train_data.Appliances.values
forest=RandomForestClassifier(warm_start=True, oob_score=True,
                               max_features="Appliances",
                               random_state=20201107),
OOB_error = []
min_estimators = 50
max_estimators = 250
for i in range(min_estimators, max_estimators + 1):
        forest=RandomForestClassifier(warm_start=True, oob_score=True,n_estimators=i,max_features="auto",
                               random_state=20201107),
        forest[0].fit(X, y)
        # Record the OOB error for each `n_estimators=i` setting.
        oob_error = 1 - forest[0].oob_score_
        OOB_error.append((i, oob_error))
````

````python
# Plot the OOB error per number of trees
plt.xlim(min_estimators, max_estimators)
plt.xlabel("n_estimators")
plt.ylabel("OOB error rate")
x_val = [x[0] for x in OOB_error]
y_val = [x[1] for x in OOB_error]
plt.plot(x_val,y_val)
plt.plot(x_val,y_val,'or')
plt.show()
````

**Written answer:**
The optimal number of trees is 110, since it gives the lowest OOB error rate, although from 200-250 they seems to give a similar or better OOB error rate, 110 is way more computational-friendly comparing with 200-250.

````python
# Final forest
final_forest=RandomForestClassifier(warm_start=True, oob_score=True,n_estimators=110,max_features="auto",
                               random_state=20201107),
# Calculate error over test set
X_test = test_data.drop(['Appliances'], axis='columns').values
y_test = test_data.Appliances.values
final_forest[0].fit(X_test,y_test)
OOB_score_final = 1-final_forest[0].oob_score_
print(OOB_score_final)
# Load the second dataset
df2 = pd.read_csv('~/Desktop/CS4414/energy_appliances_extrapolation.csv')
train_data2,test_data2 = train_test_split(df2,test_size=0.3,random_state = 20201107)
X2 = train_data2.drop(['Appliances'], axis='columns').values
y2 = train_data2.Appliances.values
# Calculate the error over it
final_forest2=RandomForestClassifier(warm_start=True, oob_score=True,n_estimators=110,max_features="auto",
                               random_state=20201107),
final_forest2[0].fit(X2,y2)
OOB_score_final2 = 1-final_forest2[0].oob_score_
print(OOB_score_final2)
````

````python
# Print MAPE over the sets
#I AM USING MACOS AND FOR SOME REASON I AM NOT ABLE TO UPDATE THE SKLEARN PACKAGE#
#I AM USING MACOS AND FOR SOME REASON I AM NOT ABLE TO UPDATE THE SKLEARN PACKAGE#
#I AM USING MACOS AND FOR SOME REASON I AM NOT ABLE TO UPDATE THE SKLEARN PACKAGE#
y_pred = final_forest[0].predict(X_test)
y_pred2 = final_forest2[0].predict(X2)
#BUT I HAVE WRITTEN THE MAPE FUNCTION JUST HOPING IT CAN RUN ON YOUR SIDE#
#BUT I HAVE WRITTEN THE MAPE FUNCTION JUST HOPING IT CAN RUN ON YOUR SIDE#
#BUT I HAVE WRITTEN THE MAPE FUNCTION JUST HOPING IT CAN RUN ON YOUR SIDE#
mape = mean_absolute_percentage_error(y_test,y_pred)
mape2 = mean_absolute_percentage_error(y2,y_pred2)
print(mape)
print(mape2)
````

**Written answer part 1 (2 pts each):**
 Why do we say we want to use a small learning rate? 
 
 A small learning rate can prevent the overfitting of the data, if the learning rate is set to be large, then the model will be fitted very closely with the training set data.
 
 Why do we say the number of trees to use depends on each dataset/problem? 
 
 The more trees used, the better the result, however, some small or simple dataset does not require a huge forest, a small number of trees is enough.
 
 Why do we want to use a small tree depth?
 
 A small tree depth will have a poor performance comparing with a deep tree and they are generally defined as weak learner, however, since we are using a boosting algorithm, we are expected to use weak learners, trees with few layers.

````python
# Define the XGB model
XGB = XGBRegressor(max_depth=3,                 # Depth of each tree
                            learning_rate=0.1,            # How much to shrink error in each subsequent training. Trade-off with no. estimators.
                            n_estimators=100,             # How many trees to use, the more the better, but decrease learning rate if many used.
#                             verbosity=1,                  # If to show more errors or not.
#                             objective='binary:logistic',  # Type of target variable.
#                             booster='gbtree',             # What to boost. Trees in this case.
#                             n_jobs=2,                     # Parallel jobs to run. Set your processor number.
#                             gamma=0.001,                  # Minimum loss reduction required to make a further partition on a leaf node of the tree. (Controls growth!)
#                             subsample=1,                  # Subsample ratio. Can set lower
#                             colsample_bytree=1,           # Subsample ratio of columns when constructing each tree.
#                             colsample_bylevel=1,          # Subsample ratio of columns when constructing each level. 0.33 is similar to random forest.
#                             colsample_bynode=1,           # Subsample ratio of columns when constructing each split.
#                             reg_alpha=1,                  # Regularizer for first fit. alpha = 1, lambda = 0 is LASSO.
#                             reg_lambda=0,                 # Regularizer for first fit.
#                             scale_pos_weight=1,           # Balancing of positive and negative weights.
#                             base_score=0.5,               # Global bias. Set to average of the target rate.
                            random_state=20201107,        # Seed
                            missing=None                  # How are nulls encoded?
                            )
````

````python
# Define the parameter grid.
param_grid = dict({'n_estimators': [350, 400, 450,500],
                   'max_depth': [3,4,5,6,7],
                 'learning_rate' : [0.01, 0.1, 0.2]
                  })
````

````python
# Create reduced validation set.
val_train = df.sample(frac = 0.3,               # The fraction to extract
                    random_state = 20201107    # The seed.
                    )
````

````python
# Train grid search. This takes a while! Go grab a coffee.
GridXGB = GridSearchCV(XGB,        # Original XGB. 
                       param_grid,          # Parameter grid
                       cv = 3,              # Number of cross-validation folds.  
#                        scoring = 'roc_auc', # How to rank outputs.
                       n_jobs = -1,         # Parallel jobs. -1 is "all you have"
                       refit = False,       # If refit at the end with the best. We'll do it manually.
                       verbose = 1          # If to show what it is doing.
                      )
X_testXGB = val_train.drop(['Appliances'], axis='columns').values
y_testXGB = val_train.Appliances.values
GridXGB.fit(X,y)
````
